```python
import numpy as np
import pandas as pd
```

```python
#가장 팁이 높은 데이터 3개의 정보를 구하세요
tips.sort_values(by='tip', ascending=False).head(3)
```

```python
#tips_pct를 구해서 tipdf에 새로운 컬럼명으로 추가하세요 
tips.loc[: ,'tips_pct'] = tips.loc[:, 'tip'] / tips.loc[:, 'total_bill']
tips
```

```python
# 팁 비율이 높은 
# 상위 5개의 정보를 출력하세요.
tips.sort_values('tips_pct', ascending=False).head()
```

```python
# 팁 비율이 낮은 
# 상위 5개의 정보를 출력하세요.
tips.sort_values('tips_pct', ascending=True).head()
```

```python
# smoker 컬럼의 값이 Yes인 데이터를 추출하고 
# 추출된 데이터 프레임에서 성별 컬럼의 고유값별로 데이터 개수를 Count 하세요.
tips.loc[tips['smoker'] == 'Yes', 'sex'].value_counts()
```

```python
# time 컬럼의 값이 Dinner인 데이터를 추출하고 
# 추출된 데이터 프레임에서 day 컬럼의 고유값별로 데이터 개수를 Count 하세요.
tips.loc[tips['time'] == 'Dinner', 'day'].value_counts()
```
