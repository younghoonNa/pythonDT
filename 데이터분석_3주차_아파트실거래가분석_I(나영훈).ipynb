```python
# loc_code.xlsx 파일을 읽어 locCode 변수에 저장하세요.
locCode = pd.read_excel('loc_code.xlsx')
locCode.head()
```

```python
# 위 데이터 중 서울특별시에 소속된 데이터만 추출하세요
# 추출은 시도명, 시군구명, 지역코드로 한정하세요
loc_Seoul = locCode.loc[locCode['시도명'] == '서울특별시', :]
loc_Seoul
```

# 2. 실거래가 데이터 가공

```python
# 2020-seoul-apt-price.csv 파일을 읽어 apt 변수에 저장하세요. encoding 옵션은 'cp949'로 설정합니다.
apt = pd.read_csv('2020-seoul-apt-price.csv', encoding = 'cp949')

apt.head()
```

```python
apt
```

```python
# 실거래가 데이터의 컬럼명을 출력하세요
apt.columns
```

```python
# 데이터의 크기를 조회하세요 (shape)
apt.shape
```

```python
# 각 열 별로 누락 데이터가 있는지 확인하세요
apt.isna().sum()
```

```python
# 누락 데이터를 제거하세요.(누락데이터가 하나라도 포함된 행은 삭제합니다.)
apt = apt.dropna()
#apt.dropna(inplaec=True)
apt.isna().sum()
```

```python
# object 타입인 '거래금액' 열을 정수 (int64)로 변경하세요
# 거래금액 컬럼의 ,를 제거 후 정수로 변경하세요.
# df['컬럼명'].str.replace(변경전, 변경후)
apt['거래금액'] = apt['거래금액'].str.replace(',', '').astype('int64')
```

```python
# 데이터의 정보를 출력하여 '거래금액' 열의 타입이 변경됐는지 확인하세요
apt.info()
```

```python
# apt['거래금액'] = pd.to_numeric(apt['거래금액'], errors = 'ignore')
# apt.info()
```

```python
# 데이터의 정보를 출력하여 '거래금액' 열의 타입이 변경됐는지 확인하세요
apt.head()
```

```python
# # 데이터의 정보를 출력하여 '거래금액' 열의 타입이 변경됐는지 확인하세요 >> 위에서 해결
# apt['거래금액'] = pd.to_numeric(apt['거래금액'], errors = 'ignore')
# apt.info()
```

```python
# '전용면적' 열을 이용하여 '전용면적(평)' 열을 추가하세요
# 변환 공식: 전용면적/3.306
# '전용면적(평)'의 값은 round 함수를 활용하여 소수점 둘째자리까지만 표기합니다.
# Ex) round(아파트 전용면적컬럼 데이터, 반올림자리수 )

apt['전용면적(평)'] = round(apt['전용면적'] , 2)
apt.head()
```

```python
# 아래 코드의 <조건> 부분을 완성하여(불리언 인덱싱),
# '거래월'이 7월 전이면 상반기, 
# 7월 이후면 하반기로 설정되는 '반기' 열을 추가하세요
apt['반기'] = 0
apt.loc[apt['거래월'] < 7 , '반기'] = "상반기"
apt.loc[apt['거래월'] >= 7 , '반기'] = "하반기"
```

```python
apt.head()
```

```python
# '전용면적(평)' 열을 이용하여 '평당거래금액' 열을 생성합니다.
apt['평당거래금액'] = apt['거래금액'] /apt['전용면적(평)']
apt.head()
```

# 3. 데이터 프레임 결합

```python
# '지역코드'를 기준으로 겹치는 지역 코드만 고려하여 법정동 코드 데이터 프레임과 
# 아파트 실거래가 데이터 프레임을 결합합니다.

df = pd.merge(apt, loc_Seoul, on = '지역코드',how = 'inner') 
df.head()
```

```python
# '지역코드' 열은 더이상 필요 없으므로 삭제합니다.
df = df.drop('지역코드', axis = 1)
df.head()
```

```python
# 완성된 데이터의 재사용성을 위해 파일로 저장합니다.
# 파일 이름은 '2020년 서울시 아파트 실거래가_merge.csv',
# encoding은 'cp949'
# index는 제외하고 저장하세요.
df.to_csv('2020년 서울시 아파트 실거래가_merge.csv', index=False, encoding='cp949')
```

# 4. 데이터 분석

```python
# 2020년에 거래된 평당거래금액이 높은 5개의 거래를 출력합니다.
df.sort_values('평당거래금액', ascending=False).head()
```

```python
# 불리언 인덱싱을 사용해서 2020년 상반기 데이터를 뽑아내고,
# 거래 금액이 높은 5개의 거래를 출력합니다.
df.loc[df['반기'] == '상반기', :].sort_values('거래금액').tail()
```

```python
# groupby를 사용하여, 시군구 별 거래 금액의 합계와 평균 금액을 구하고
# 거래 금액의 평균이 높은 순서대로 정렬합니다.
pd.DataFrame(df.groupby('시군구명')['거래금액'].agg(['sum', 'mean'])).sort_values('mean', ascending=False)
```

```python
# 거래월과 시군구별 거래량(거래 건 수)을 피벗테이블로 표현하세요
pv_table = df.pivot_table(index='시군구명', values='거래금액',  columns='거래월', aggfunc='size')
pv_table
```

```python
# 시군구가 강남구인 데이터를 추출하세요
gangnam = df.loc[df['시군구명'] == '강남구' , :]
gangnam
```

```python
# 강남구의 아파트 단지별 평균 거래 금액을 계산 후 상위 5개를 출력합니다.
pd.DataFrame(gangnam.groupby(['아파트이름'])['거래금액'].mean()).sort_values('거래금액', ascending=False).head()
```

```python
# 피벗 테이블 이용하여 강남구의 거래월과 법정동의 평균거래금액을 계산합니다. (NA는 0으로 치환합니다.)
# 완성된 피벗 테이블 뒤에
# pivot_table( ... ).round(반올림자리수)
# 형태로 소수점 둘째자리까지 출력합니다.
# 거래월과 시군구별 거래량(거래 건 수)을 피벗테이블로 표현하세요
pv_gang_table = gangnam.pivot_table(values='거래금액', index='법정동', columns='거래월', aggfunc = 'mean').fillna(0).round(2)
pv_gang_table
```
