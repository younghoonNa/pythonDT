```python
x_train, x_test, y_train, y_test  = train_test_split(X,Y,random_state = 0)

k_range = range(1,101)

train_acc = []
test_acc = []

for k in k_range:
  model = KNeighborsClassifier(n_neighbors=k).fit(x_train, y_train)
  train_acc.append(model.score(x_train, y_train))
  test_acc.append(model.score(x_test, y_test))
```

```python
# 실험 결과 시각화
plt.figure(figsize=(30,15))
plt.plot(k_range, train_acc, label='Train Accuracy')
plt.plot(k_range, test_acc, label='Test Accuracy')
plt.legend()
plt.grid()
```

```python
# 실험 결과 시각화
plt.figure(figsize=(30,15))
plt.plot(k_range[:20], train_acc[:20], label='Train Accuracy')
plt.plot(k_range[:20], test_acc[:20], label='Test Accuracy')
plt.legend()
plt.grid()
plt.xticks(k_range[:20])
plt.show()
```

```python
x_train, x_test, y_train, y_test  = train_test_split(X,Y,random_state = 1)

k_range = range(1,101)

train_acc = []
test_acc = []

for k in k_range:
  model = KNeighborsClassifier(n_neighbors=k).fit(x_train, y_train)
  train_acc.append(model.score(x_train, y_train))
  test_acc.append(model.score(x_test, y_test))
```

```python
# 실험 결과 시각화
plt.figure(figsize=(30,15))
plt.plot(k_range, train_acc, label='Train Accuracy')
plt.plot(k_range, test_acc, label='Test Accuracy')
plt.legend()
plt.grid()
```

```python
# 실험 결과 시각화
plt.figure(figsize=(30,15))
plt.plot(k_range[:20], train_acc[:20], label='Train Accuracy')
plt.plot(k_range[:20], test_acc[:20], label='Test Accuracy')
plt.legend()
plt.grid()
plt.xticks(k_range[:20])
plt.show()
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
df['target'] = boston['target']
df
```

```python
# 상관관계 확인 > 선형회귀 (독립성) > 다중공선성
  # 독립성 : 특성간의 상관관계가 없어야 한다 (0.9 미만)
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), fmt='.2f', annot=True, cmap='Blues')
plt.show()
```

```python
X = df.loc[:, 'CRIM':'LSTAT']
Y = df.loc[:, 'target']

from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(X,Y, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression  #회기모델, 지도학습, 선형모델
model = LinearRegression().fit(x_train, y_train)
model.score(x_train, y_train) , model.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
p_test = model.predict(x_test)
mean_absolute_error(y_test, p_test)
#MAE는 약 3.6 단위가 k 이므로 3600
```

```python
#선형성 - 파란 점은 잔차의 차이를, 빨간색은 잔차의 추이를 나타낸다.
# 잔차는 0에 가까울 수록 좋다. 따라서 회색 점선에 얼마나 가까운지를 보며 선형성을 확인할 수 있다.
residual = y_test - p_test
sns.regplot(p_test, residual, lowess=True, line_kws={'color':'red'})
plt.plot([p_test.min(), p_test.max()], [0,0], '--', color='gray')
```

```python
#정규성
import scipy.stats
sr = scipy.stats.zscore(residual)
(x,y) , _ = scipy.stats.probplot(sr)
sns.scatterplot(x,y)
plt.plot([-3, 3], [-3, 3], '--', color='gray')
```

```python
#등분산성 -> 잔차들의 분산이 모여있는지 확인.
sns.regplot(p_test, np.sqrt(np.abs(sr)), lowess=True, line_kws={'color':'red'})
```
