```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.arange(10)
plt.plot(x)
```

```python
y = x ** 2
print(x)
print(y)
plt.plot(x, y)
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
```

```python
x = np.arange(10)
y = x **2
plt.plot(x,y)
plt.xlim(2,5)
plt.ylim(5,20)
```

```python
y2 = x * 5 + 2
plt.plot(x,y, label = 'first')
plt.plot(x,y2, label = 'second')
plt.legend()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('total_bill')
plt.ylabel('tip')
```

```python
res = tips.groupby('day')['tip'].mean()
plt.bar(res.index, res.values)
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(data = tips, y='total_bill', x = 'day')
```

```python
sns.boxplot(data=tips, y='total_bill', x='day', hue='smoker') 
```
