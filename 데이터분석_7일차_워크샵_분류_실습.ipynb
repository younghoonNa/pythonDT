```python
# 학습에 사용할 특성을 선택하세요. (학습에서 제외하기 위해선 주석처리 합니다.)
features = [
     'V1',
     'V2',
     'V3',
     'V4',
     'V5',
     'V6',
     'V7',
     'V8',
     'V9',
     'V10',
     'V11',
     'V12',
     'V13',
     'V14',
     'V15',
     'V16',
     'V17',
     'V18',
     'V19',
     'V20',
     'V21',
     'V22',
     'V23',
     'V24',
     'V25',
     'V26',
     'V27',
     'V28',
     'Amount_log'
]
features
```

```python
# features에 저장된 컬럼명만 추출하여 X에 대입하세요.
X = df[features]
X
```

```python
# Class 컬럼을 추출하여 Y에 대입하세요
Y = df['Class']
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import classification_report
```

```python
# 추출된 X데이터를 RobustSclaer를 이용하여 정규화하세요.
scaler = RobustScaler()
x_scaled = scaler.fit_transform(X)
x_scaled
```

```python
# X_Sclaed와 Y를 이용하여 데이터를 분할하세요. 
# 층화추출을 적용하세요, random_state는 0으로 지정하세요.
x_train, x_test, y_train, y_test = train_test_split(x_scaled, Y, 
                                                    stratify=Y,
                                                    test_size=0.01,
                                                    random_state=0)
```

```python
# KNN 모델을 적용 후 성능을 평가하세요. 
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier().fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
