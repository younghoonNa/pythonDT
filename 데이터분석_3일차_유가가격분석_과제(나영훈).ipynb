```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.rc('font', family='NanumBarunGothic') 
```

# 1. 데이터 불러오기

```python
first = pd.read_csv('gas_first_2019.csv', encoding='euc-kr')
second = pd.read_csv('gas_second_2019.csv', encoding='euc-kr')
```

# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first, second], axis=0, ignore_index=True)
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
df.sort_values('휘발유', ascending = False).head()
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
df.loc[df['휘발유'] > 0, :].sort_values('휘발유').head()
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
df_drink = df.groupby(['상표'])['휘발유'].mean().reset_index()
# df_drink = pd.DataFrame(df_drink)
df_drink
```

```python
df_drink['상표']
```

```python
colors = sns.color_palette('hls',len(df_drink)) ## 색상 지정

#참조 스택오버플로우 https://stackoverflow.com/questions/18973404/setting-different-bar-color-in-matplotlib-python
#https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html
```

```python
# 상표별 평균 휘발유 가격을 시각화하세요.
plt.figure(figsize=(10, 7))
plt.bar(df_drink['상표'], df_drink['휘발유'], color = colors)
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
df_self = df.groupby(['상표', '셀프여부'])['휘발유'].mean()
df_self = pd.DataFrame(df_self)
df_self
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 시각화하세요.
plt.figure(figsize=(20, 15))
colors = sns.color_palette('hls',len(df_self)) ## 색상 지정

names = ['GS칼텍스 셀프', 'GS칼텍스 일반', 'NH-OIL 일반', 'S-OIL 셀프', 'S-OIL 일반', 'SK에너지 셀프', 'SK에너지 일반', '알뜰(ex) 셀프', 
            '알뜰주유소 셀프', '알뜰주유소 일반', '자가상표 셀프', '자가상표 일반', '현대오일뱅크 셀프', '현대오일뱅크 일반']
plt.bar(names, df_self['휘발유'],  color = colors)
```

#다시 시작 <br>
0을 가진 휘발유 가격 제거 한 상태로 진행하셔 강사님 답과 다를 수 있습니다.

```python
# 지역별 평균 휘발유 가격을 계산하세요
locate = df_notzero_OIL.groupby(['구'])['휘발유'].mean().reset_index()
locate
```

```python
# 지역별 평균 휘발유 가격을 시각화하세요.
colors = sns.color_palette('hls',len(locate)) ## 색상 지정
plt.figure(figsize=(25, 20))
plt.bar(locate['구'], locate['휘발유'], color = colors)
```

```python
# pivot_table 함수를 이용하여 구와 상표 컬럼을 이용하여 평균 휘발유 가격을 계산하세요
locate_brand = df_notzero_OIL.groupby(['구', '상표'])['휘발유'].mean().reset_index()

locate_p = locate_brand.pivot('구', '상표', '휘발유')
locate_p
```

```python
# pivot_table 함수를 이용하여 구와 월 컬럼을 이용하여 평균 휘발유 가격을 시각화 하세요 (heatmap 이용)
plt.figure(figsize=(10, 7))
sns.heatmap(locate_p)
```
